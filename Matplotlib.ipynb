```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel("numbers")
plt.xlabel("indices")
plt.title('MyPlot')
plt.show()
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])  #plots (1,1)|(2,4) and so on
plt.grid() #enables grid
plt.xlabel("numbers")
plt.ylabel("squares")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')  # r->red & o-> o shaped dot
plt.grid()
plt.show()
```

```python
import numpy as np
a = np.arange(0.,5.,0.2)

plt.plot(a,a**2, 'b--', label='^2')   #b-- => blue as --
plt.plot(a,a**2.2,'rs', label='^2.2') #rs => red as square
plt.plot(a,a**2.5,'y^', label='^2.5') #y^ => yellow as triangle

plt.grid()
plt.legend()
plt.show()
```

**Controlling line properties**

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y,linewidth=5.0)
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]

x2 = [1,2,3,4]
y2 = [2,4,6,8]

lines = plt.plot(x1,y1, x2,y2) #x1,y1 is 0 and x2,y2 is 1

#using python args keyword
plt.setp(lines[0],color='r',linewidth=2.0)

#using MATLAB style string value pair
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)

plt.grid()
plt.show()
```

## **Multiple Axes and Figures**

```python
def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1) #make a figure

plt.subplot(211) #figure me 2 rows, 1 column ka 1st part me ye subplot hoga
plt.plot(t1,f(t1), 'b-')
plt.grid()

plt.subplot(212) #figure me 2 rows, 1 column ka 2nd part me ye subplot hoga
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```

```python
plt.figure(1)             #first figure
plt.subplot(211)          #first subplot in the first figure
plt.plot([1,2,3])
plt.subplot(212)          #second subplot in the first figure
plt.plot([4,5,6])


plt.figure(2)             #second figure
plt.plot([7,8,9])         #automatically creates a subplot(111)


plt.figure(1)             #figure 1| subplot(212) is current
plt.subplot(211)          #make subplot(211) as current
plt.title('Easy 123')     #subplot 211 title

plt.show()
```
